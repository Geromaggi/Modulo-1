```python
archivo = open('hospitales2.csv',encoding='utf-8')
tabla = csv.reader(archivo)
with open('hospitales_salida.csv','w',encoding='utf-8',newline='') as salida: #se pone el newline por el salto de linea
    salida_writter = csv.writer(salida)
    next(tabla)
    salida_writter.writerow(['latitude','longitude','name','label'])
    for linea in tabla:
        coordenadas = linea[0][7:-1].split()
        salida_writter.writerow([coordenadas[1],coordenadas[0],linea[8],linea[3]])
archivo.close()
```
