```python
def dec_bin (numero_decimal):
    if (type(numero_decimal) != float):
        return None
    elif (numero_decimal <= 0):
        return None
    else:
        dec = 24
        lista_binario = []
        while numero_decimal > 0 and dec > 0:
            lista_binario.append(int(numero_decimal * 2))
            numero_decimal = numero_decimal * 2 - int(numero_decimal * 2)
            dec -= 1
        numero_binario = '0.'

        for num in lista_binario:
            numero_binario += str(num)
        return numero_binario
```
