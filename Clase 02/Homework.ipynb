```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
separador = "|"
with open ('Emisiones_CO2.csv',encoding='latin-1') as archivo:
    next(archivo)
    for linea in archivo:
        linea = linea.rstrip("\n") #se puede poner:
        #linea = linea.strip()  lo que hace esto es que no salga el /n que siempre esta al final de cada linea de cualquier archivo. 
        columnas = linea.split(separador) 
        
        codigo_pais = columnas[0]
        dicc_emisiones['cod_pais'].append(codigo_pais) # de manera mas compacta puede hacerse asi:
        #dicc_emisiones['cod_pais'].append(columnas[0])
        
        nombre_pais = columnas[1]
        dicc_emisiones['nom_pais'].append(nombre_pais)

        region = columnas[2]
        dicc_emisiones['region'].append(region)
        
        anio = int(columnas[3])
        dicc_emisiones['anio'].append(anio)

        co2_kt = columnas[4]
        dicc_emisiones['co2'].append(co2_kt)

        co2_per_capita = columnas[5]
        dicc_emisiones['co2_percapita'].append(co2_per_capita)
    #print(dicc_emisiones) 
```

e) ¿Cuál es el total de emisiones de CO2 para 'América Latina y Caribe' en el año 2010?

```python
def normalizacion (lista):
    for indice,elemento in enumerate(lista):
        if elemento:
            elemento = elemento.replace('.','')
            elemento = elemento.replace(',','.')
            elemento = float(elemento)
        else: elemento = None
        lista[indice] = elemento
    return lista
```

```python
dicc_emisiones['co2'] = normalizacion(dicc_emisiones['co2'])
dicc_emisiones['co2_percapita'] = normalizacion(dicc_emisiones['co2_percapita'])
```

```python
for i in range (len(dicc_emisiones['anio'])):
    dicc_emisiones['anio'][i] = int(dicc_emisiones['anio'][i])
```

```python
filtro_region = 'América Latina y Caribe'
filtro_anio = 2010
emisiones = 0
for indice,elemento in enumerate(dicc_emisiones['region']):
    if dicc_emisiones['region'][indice] == filtro_region and dicc_emisiones['anio'][indice] == filtro_anio and dicc_emisiones['co2'][indice] != None:
        emisiones += dicc_emisiones['co2'][indice]
print(f'Las emisiones en {filtro_region} en {filtro_anio} fueron de {round(emisiones,3)} kt') #con round redondeamos a los digitos que le pasemos como segundo parametro. 
```
